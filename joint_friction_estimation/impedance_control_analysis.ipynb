```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 7)
```

```python
# filename = "finger0_PD_follow_square_09-Jul-2024_19_15_32.npy"
filename = "finger0_PD_follow_square_09-Jul-2024_19_19_19.npy"
data = np.load(filename, allow_pickle=True).item()
```

```python
%matplotlib widget
fig, ax = plt.subplots(3)

ax[0].plot(data["fingertip_cur_pos"][:, 0], c="r")
ax[0].plot(data["fingertip_desired_pos"][:, 0], c="b")
ax[0].legend(["Actual x", "Desired x"], loc="upper right")
ax[0].set_ylabel("x (m)")
ax[0].set_xlabel("Time (ms)")

ax[1].plot(data["fingertip_cur_pos"][:, 1], c="r")
ax[1].plot(data["fingertip_desired_pos"][:, 1], c="b")
ax[1].legend(["Actual y", "Desired y"], loc="upper right")
ax[1].set_ylabel("y (m)")
ax[1].set_xlabel("Time (ms)")

ax[2].plot(data["fingertip_cur_pos"][:, 2], c="r")
ax[2].plot(data["fingertip_desired_pos"][:, 2], c="b")
ax[2].legend(["Actual z", "Desired z"], loc="upper left")
ax[2].set_ylabel("z (m)")
ax[2].set_xlabel("Time (ms)")

fig.suptitle("Fingertip position vs desired position (Impedance Controller)")
plt.tight_layout()
plt.show()
# plt.savefig("fingertip_vs_desired_position.png", dpi=600)
```
